```python
import matplotlib.pyplot as plt
import pandas as pd

# Leer el archivo CSV
df = pd.read_csv("Popular_Spotify_Songs.csv", encoding="latin1")

df = df.drop(['track_name', 'artist(s)_name', 'artist_count', 'released_year', 'released_month', 
              'released_day', 'in_spotify_playlists', 'in_spotify_charts', 'in_apple_charts', 
              'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'streams', 'in_apple_playlists', 
              'key', 'mode', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%'], axis=1)

valores_nulos = df.isnull().sum()

# Mostrar los valores nulos
print(valores_nulos)

```

```python
# Calcular las medias de danceability, valence y energy para cada rango de BPM
danceability_moderate = df[(df['bpm'] >= 0) & (df['bpm'] < 90)]['danceability_%'].mean()
danceability_fast = df[df['bpm'] >= 90]['danceability_%'].mean()

valence_moderate = df[(df['bpm'] >= 0) & (df['bpm'] < 90)]['valence_%'].mean()
valence_fast = df[df['bpm'] >= 90]['valence_%'].mean()

energy_moderate = df[(df['bpm'] >= 0) & (df['bpm'] < 90)]['energy_%'].mean()
energy_fast = df[df['bpm'] >= 90]['energy_%'].mean()

# Graficar
plt.figure(figsize=(10, 6))

# Valence
plt.bar('Danceability\nmoderate', danceability_moderate, color='#05668d', alpha=0.7, label='Moderate')

# Energy
plt.bar('Danceability\nfast', danceability_fast, color='#02c39a', alpha=0.7, label='Fast')

plt.bar(' ', danceability_fast, color='white', alpha=0.05)

# Valence
plt.bar('Valence\nmoderate', valence_moderate, color='#05668d', alpha=0.7)

# Valence
plt.bar('Valence\nfast', valence_fast, color='#02c39a', alpha=0.7)

plt.bar('  ', danceability_fast, color='white', alpha=0.05)

# Energy
plt.bar('Energy\nmoderate', energy_moderate, color='#05668d', alpha=0.7)

# Energy
plt.bar('Energy\nfast', energy_fast, color='#02c39a', alpha=0.7)

plt.xlabel('\n\nBPM range')
plt.ylabel('Average Rating')
plt.title('Relationship of tempo (BPM) with danceability, energy and valence of music')
plt.legend()
plt.show()
```
